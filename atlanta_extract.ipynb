```python
# create empty lists for data
name = []
city = []
latitude = []
longitude = []
zipcode = []
ethnicity = []
price = []
rating = []

for i in range(len(atlanta_responses)):
    for j in range(len(atlanta_responses[i]['businesses'])):
        name.append(atlanta_responses[i]["businesses"][j]["name"])
        city.append("Atlanta")
        latitude.append(atlanta_responses[i]["businesses"][j]["coordinates"]["latitude"])
        longitude.append(atlanta_responses[i]["businesses"][j]["coordinates"]["longitude"])
        zipcode.append(atlanta_responses[i]["businesses"][j]["location"]["zip_code"])
        ethnicity.append((atlanta_responses[i]['Ethnicity']))
        try:
            price.append(atlanta_responses[i]["businesses"][j]["price"])
        except KeyError:
            price.append("N/A")
        rating.append(atlanta_responses[i]["businesses"][j]["rating"])
```

```python
print(len(name))
print(len(city))
print(len(latitude))
print(len(longitude))
print(len(zipcode))
print(len(rating))
```

```python
atlanta_all_responses = {
    "name": name,
    "city": city,
    "latitude": latitude,
    "longitude": longitude,
    "zipcode": zipcode,
    "ethnicity": ethnicity,
    "price": price,
    "rating": rating}
```

```python
atlanta_restaurants = pd.DataFrame(atlanta_all_responses)
atlanta_restaurants
```
